```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.eye(5)
```

```python
path = '../MachineLearning/data/ex1data1.txt'
data = np.loadtxt(path,delimiter=',')
```

```python
plt.scatter(data[:, 0], data[:, 1], s=20, marker='x', c='r')
plt.xlabel("Population")
plt.ylabel("Profit")
plt.show()
```

```python
def computeCost(x, y, theta):
    inner = np.power((x@theta-y), 2)
    return np.sum(inner) / (2*x.shape[0])
```

```python
def updateTheta(x, y, theta, alpha):
    inner = (x @ theta - y).T @ x
    return theta - (inner * alpha / x.shape[0]).T
```

```python
x = data[:,0]
y = data[:,1]
y = y.reshape((len(y),1))
x = np.c_[np.ones(len(x)), x]
theta = np.zeros((2,1))
computeCost(x,y,theta)
```

```python
range(10)
```

```python
loss = []
for _ in range(1500):
    theta = updateTheta(x, y, theta, 0.01)
    loss.append(computeCost(x, y, theta))

plt.title('LOSS')
plt.plot(np.arange(1500), loss)
plt.xlabel('Times')
plt.ylabel('Loss')
plt.show()
```

```python
py = []
for i in range(x.shape[0]):
    py.append(x[i] @ theta)

plt.title('Predicted Profit vs. Population Size')
plt.scatter(data[:, 0], data[:, 1], s=20, marker='x')
plt.plot(data[:, 0], py, 'r', linewidth=.5)
plt.xlabel("Population")
plt.ylabel("Profit")
plt.show()
```
